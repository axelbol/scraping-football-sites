```python
import bs4
import requests
from io import StringIO
```

```python
# In this case the table is commented out

# Fetch the web page content
response = requests.get('https://fbref.com/en/comps/32/2023-2024/stats/2023-2024-Primeira-Liga-Stats')
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# Find all comments
# comments = soup.find_all(text=lambda text:isinstance(text, bs4.Comment))
comments = soup.find_all(string=lambda text:isinstance(text, bs4.Comment))
# commented_out_tables = [bs4.BeautifulSoup(cmt).find_all('table') for cmt in comments]
commented_out_tables = [bs4.BeautifulSoup(cmt, 'html.parser').find_all('table') for cmt in comments]

# Filter to keep only single tables
commented_out_tables = [tab[0] for tab in commented_out_tables if len(tab) == 1]

# Convert the table to a DF
# df_primeira_liga = pd.read_html(str(commented_out_tables[0]))[0]
df_primeira_liga = pd.read_html(StringIO(str(commented_out_tables[0])))[0]
```
