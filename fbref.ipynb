**FBREF** (https://fbref.com)

Fbref is a popular website for advanced statistics and advanced data for football analysis. Fbref is build in tables so it is easy to scrape. We are going to use css selectors and pandas.

```python
import pandas as pd
```

```python
# we'll start off by scraping the data from the 2023/2024 Champions League
df_fbref_champions = pd.read_html('https://fbref.com/en/comps/8/2023-2024/2023-2024-Champions-League-Stats', attrs={'id': 'results2023-202480_overall'})[0]
```

```python
df_fbref_champions.head()
```

```python
# fbref uses a row for a separator, remove those rows
df_fbref_champions = df_fbref_champions.dropna(subset='Rk')
```

```python
df_fbref_champions.head(10)
```

```python
# not all tables are made equal
# Real Madrid 23-24 stats
df_fbref_real_madrid = pd.read_html('https://fbref.com/en/squads/53a2f082/2023-2024/Real-Madrid-Stats', attrs={'id': 'stats_standard_12'})[0]
```

```python
# The columns are multi-indexed, we'll flatten the columns
df_fbref_real_madrid.columns = ['_'.join(col).strip() for col in df_fbref_real_madrid.columns.values]
```

```python
# Sometimes the tables are commented out, so we'll have to extract the tables from the comments which is a bit more complicated
import bs4
import requests
from io import StringIO
```

```python
# In this case the table is commented out
# fetch the web page content
response = requests.get('https://fbref.com/en/comps/32/2023-2024/stats/2023-2024-Primeira-Liga-Stats')
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# find all comments
comments = soup.find_all(string=lambda text:isinstance(text, bs4.Comment))
commented_out_tables = [bs4.BeautifulSoup(cmt, 'html.parser').find_all('table') for cmt in comments]

# filter to keep only single tables
commented_out_tables = [tab[0] for tab in commented_out_tables if len(tab) == 1]

# convert the table to a DF
df_primeira_liga = pd.read_html(StringIO(str(commented_out_tables[0])))[0]
```
